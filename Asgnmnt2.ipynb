```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Reading the dataset

```python
data = pd.read_csv('cars_data.csv')
```

```python
data
```

# 2. Check for null values

# 3.Bar Graph 

```python
dat = data['Buyer Gender'].value_counts()
dat
```

```python
dat.plot(kind='bar',color=['green','red'])
plt.xticks(rotation=0)
plt.ylim(4800,5100)
plt.grid()
```

It is clear that in this dataset female buyers are more than male buyers 

# 4. Top 5 cars based on sales price 

```python
dat1 = data[[ 'Model','Sale Price']]
d = dat1.sort_values(['Sale Price'],ascending = False)
d.head()
```

From the above Jetta III, A6,LX,MX-6,ES are the top 5 cars based on sales price
